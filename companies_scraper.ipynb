```python
# Sample HTML content (replace this with your actual HTML content)
html_content = response.text

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all the items with the class "item col-12 list-group-item"
items = soup.find_all('div', class_= 'list-group-item')
company_names = []
img_urls = []
countries = []
stand_infos = []
details = []
sectors_list = []
profile_urls = []


# Iterate through each item and extract the desired information
for item in items:
    # Extract the image URL
    img_src = item.find('img', class_='group list-group-image img-fluid')['src']
    
    # Extract the company name
    company_name = item.find('h4', class_='heading').text.strip()
    
    # Extract the company details
    detail = item.find('p', class_='list-group-item-text').text.strip()
    
    # Extract the country
    country = item.find('span', style='line-height:22px; font-size: 12px; font-weight: 600; float:left;').text.strip()
    
    # Extract the stand and hall information
    stand_info = item.find('p', style='margin-bottom:0;').text.strip()
    
    # Extract the sectors
    sectors = [sector.text.strip() for sector in item.find_all('li')]
    
    # Extract the profile URL
    profile_url = item.find('a', href=True)['href']
    
    # Print or store the extracted information as needed
#     print(f"Company Name: {company_name}")
#     print(f"Image URL: {img_src}")
#     print(f"Country: {country}")
#     print(f"Stand Info: {stand_info}")
#     print(f"Details: {detail}")
#     print(f"Sectors: {sectors}")
#     print(f"Profile URL: {profile_url}")
#     print("\n")
    
    
        # Append extracted data to lists
    company_names.append(company_name)
    img_urls.append(img_src)
    countries.append(country)
    stand_infos.append(stand_info)
    details.append(detail)
    sectors_list.append(sectors)
    profile_urls.append(profile_url)
    
# Create a DataFrame from the extracted data
data = {
    'Company Name': company_names,
    'Image URL': img_urls,
    'Country': countries,
    'Stand Info': stand_infos,
    'Details': details,
    'Sectors': sectors_list,
    'Profile URL': profile_urls
}
df = pd.DataFrame(data)
```

```python
# Removing column 'B'
df = df.drop(['Image URL'], axis=1)
```

```python
# Function to scrape data from URL
def scrape_profile_data(url):
    response = requests.get(url, verify=False)
    soup = BeautifulSoup(response.content, 'html.parser')
    # Perform scraping operations here to extract the desired data from the webpage
    items = soup.find_all('li', class_= 'social_website')
    # Extract the first href link or return None if not found
    #href_link = next((item.find('a')['href'] for item in items if item.find('a')), None)
    
    
    href_links = [item.find('a')['href'] for item in items]
    
    #href_links = item.find('a', href=True)['href']
    
    
    return href_links[0]
```

```python
# Iterate through 'Profile URL' column, scrape data, and append to DataFrame
for index, row in df.iterrows():
    profile_url = row['Profile URL']
    scraped_data = scrape_profile_data(profile_url)
    df.at[index, 'companay website'] = scraped_data

# Print updated DataFrame
# print(df)
```
