```python
#create a function to clean tweets
def cleanTxt(text):
    text = text.replace('\u2026','')
    text = re.sub(r'@[A-Za-z0-9_:]+', '', text) #remove @mentions
    text = re.sub(r'RT[\s]+', '', text) # remove ReTweets
    text = re.sub(r'https?:\/\/\S+', '', text) # remove hyperlinks 
    text = re.sub(r'(\n)', '', text)
    text = text.lower()
    
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    text = re.sub(r'[!]+', '', text)  # remove exclamation marks
    # remove emoticons
    text = re.sub(r'[:;=][\)\(\[\]DPOp]([pPoODsS/\\]+[vV])?|[\)\(\[\]DPOp][:=;][vV]?|[Xx][:;=][\)\(\[\]DPOp]|<3', '', text)
    # Remove special characters and symbols
    text = ''.join(c for c in text if unicodedata.category(c) != 'So' and unicodedata.category(c) != 'Sk')
    # Remove additional patterns
    text = re.sub(r'\[[^\]]*\]', '', text)  # remove square brackets and their contents
    text = re.sub(r'â€™|â€˜|â€œ|â€', '', text)  # remove specific quotation marks
    text = re.sub(r'â€”|â€“', '', text)  # remove dashes
    text = re.sub(r'\u2066|\u2069|\u200d', '', text)  # remove specific Unicode characters
    text = re.sub(r'Â¡|Â¢|â€¼ï¸|â€¢', '', text)  # remove specific symbols
    text = re.sub(r'ðŸ®ðŸ¬ðŸ®ðŸ¯|ï¸âƒ£', '', text)  # remove specific Unicode characters
    # remove specific Chinese text
    text = re.sub(r'æˆ‘å’Œæˆ‘çš„å›¢é˜Ÿä¸ŽåŠ å·žç†å‘å’Œç¾Žå®¹å§”å‘˜ä¼šåˆä½œï¼Œæä¾›ç®€ä½“ä¸­æ–‡è€ƒè¯•ï¼æˆ‘ä»¬çš„ç›®æ ‡æ˜¯ä¸ºç†å‘å’Œç¾Žå®¹è¡Œä¸šçš„å­¦ç”Ÿå’Œè¢«è®¸å¯äººæä¾›æ›´å¥½çš„ç¤¾åŒºè®¿é—®å¹¶å‡å°‘è¯­è¨€éšœç¢ã€‚æˆ‘ä»¬å¾ˆè‡ªè±ªåœ°å®£å¸ƒè¿™äº›è€ƒè¯•çŽ°å·²ç”Ÿæ•ˆï¼Œæ‰€ä»¥è¯·åˆ†äº«è¿™ä¸ªæ¶ˆæ¯', '', text)
    text = re.sub(r'æˆ‘å‘è’™ç‰¹åˆ©å…¬å›­', '', text)  # remove specific Chinese text
    text = re.sub(r'æžªå‡»æ¡ˆçš„å—å®³è€…å’Œä»–ä»¬çš„å®¶äººè¡¨ç¤ºå“€æ‚¼ã€‚è’™ç‰¹åˆ©å…¬å›­å…°åˆ©è€å¹´ä¸­å¿ƒ', '', text)  # remove specific Chinese text
    text = re.sub(r'æä¾›å—å®³è€…æœåŠ¡ã€‚è¯·åˆ°é‚£é‡Œå¯»æ±‚å¸®åŠ©', '', text)  # remove specific Chinese text
    text = re.sub(r'æ–°å¹´å¿«ä¹', '', text)  # remove specific Chinese text
    text = re.sub(r'\u200dfunding', '', text)  # remove specific Unicode characters
    text = re.sub(r'ï¸\u200dï¸\u200dï¸thank', '', text)  # remove specific Unicode characters
    text = re.sub(r'\u2066to', '', text)  # remove specific Unicode characters
    text = re.sub(r'\u2066', '', text)  # remove specific Unicode characters
    text = re.sub(r'abâ€“', '', text)  # remove specific text
    text = re.sub(r'Â¡ha', '', text)  # remove specific text
    text = re.sub(r'Â¡feliz', '', text)  # remove specific text
    text = re.sub(r'Â¡si', '', text)  # remove specific text
    text = re.sub(r'\u2066latest', '', text)  # remove specific Unicode characters
    text = re.sub(r'â†“|â†’', '', text)  # remove arrows
    text = re.sub(r'backsâ€”and|backâ€”now', '', text)  # remove specific phrases
    text = re.sub(r'â€¼ï¸', '', text)  # remove specific Unicode characters
    text = re.sub(r'â€¢|â€¢transitional|â€¢national', '', text)  # remove specific phrases
    text = re.sub(r"['\[\],\u200b]", "", text)
    text = re.sub(r"['\[\],\u200b\uFE0F\u200D]", "", text)
    text = re.sub(r'[\[\]]', '', text)
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = re.sub(r'\bu\b|\bca\b', '', text)
    return text
```

## Stopwords

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
#english_alphabet = [chr(x) for x in range(ord('a'), ord('z')+1)]
```

```python
stop_words
```

```python
def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
    return ' '.join(filtered_tokens)
```

```python
df_fl['stopwords_text'] = df_fl['tokenized_text'].apply(remove_stopwords)
df_Ca['stopwords_text'] = df_Ca['tokenized_text'].apply(remove_stopwords)
```

## Lematization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text)
    filtered_tokens = [lemmatizer.lemmatize(token.lower()) for token in tokens if token.lower() not in stop_words]
    return ' '.join(filtered_tokens)
```

```python
df_fl['lemma_text'] = df_fl['stopwords_text'].apply(preprocess_text)
df_Ca['lemma_text'] = df_Ca['stopwords_text'].apply(preprocess_text)
```

## Elbow Method

```python
# Realizar el mÃ©todo del codo para seleccionar el nÃºmero Ã³ptimo de clusters
num_clusters = range(1, 40)
inertia_values = []


for k in num_clusters:
    kmeans = KMeans(n_clusters=k, n_init=10,random_state=42)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

# Visualizar el mÃ©todo del codo
plt.plot(num_clusters, inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inercia')
plt.title('Elbow Method')
plt.show()
```

```python
mycenters = pd.DataFrame({'Clusters' : k, 'WSS' : inertia_values})
mycenters
```

## K-means

```python
# Configurar y ejecutar el algoritmo de k-means
num_clusters = 10  # NÃºmero de clusters deseados
kmeans = KMeans(n_clusters=num_clusters, random_state=123)
kmeans.fit(X)

# Obtener las etiquetas de los clusters
etiquetas_clusters = kmeans.labels_

# Agregar las etiquetas de clÃºster al DataFrame original
df_filter['cluster'] = etiquetas_clusters
```

```python
import pandas as pd
from collections import Counter

# Crear una lista para almacenar las palabras mÃ¡s frecuentes en cada clÃºster
cluster_word_frequency = []

# Iterar sobre cada clÃºster
for cluster_id in range(num_clusters):
    # Filtrar el DataFrame por clÃºster
    cluster_df = df_filter[df_filter['cluster'] == cluster_id]
    
    # Obtener los textos del clÃºster
    cluster_texts = cluster_df['lemma_text']
    
    # Concatenar todos los textos en un solo texto
    combined_text = ' '.join(cluster_texts)
    
    # Tokenizar el texto en palabras individuales
    tokens = combined_text.split()
    
    # Calcular la frecuencia de cada palabra
    word_frequency = Counter(tokens)
    
    # Guardar las palabras mÃ¡s frecuentes en la lista (solo las 5 principales)
    words = [word for word, frequency in word_frequency.most_common(10)]  # Obtener las 5 palabras mÃ¡s frecuentes
    cluster_word_frequency.append(words)

# Crear un DataFrame con los resultados
df_cluster_word_frequency = pd.DataFrame({'Cluster': range(num_clusters), 'Top Words': cluster_word_frequency})

# Mostrar el DataFrame completo
with pd.option_context('display.max_colwidth', None):
    print(df_cluster_word_frequency)
```
