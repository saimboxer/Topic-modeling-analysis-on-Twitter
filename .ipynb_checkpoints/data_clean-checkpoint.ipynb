```python
#clean the text

#create a function to clean tweets
def cleanTxt(text):
    text = text.replace('\u2026','')
    text = re.sub(r'@[A-Za-z0-9_:]+', '', text) #remove @mentions
    text = re.sub(r'RT[\s]+', '', text) # remove ReTweets
    text = re.sub(r'https?:\/\/\S+', '', text) # remove hyperlinks 
    text = re.sub(r'(\n)', '', text)
    text = text.lower()
    
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)

    return text
```

```python
df['text_clean'] = df['text'].apply(cleanTxt)
```

```python
from nltk.corpus import stopwords  
from nltk.tokenize import word_tokenize  
stop_words = set(stopwords.words('english'))  

def remove_stopwords(sentence):
    word_tokens = word_tokenize(sentence)  
    clean_tokens = [w for w in word_tokens if not w in stop_words]  
    
    return clean_tokens
    
df['S'] = df['text_clean'].apply(remove_stopwords)
```

```python
df['text_string'] = df['stopwords_text'].apply(lambda x: ' '.join([item for item in x if len(item)>2]))
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('omw-1.4')
lmtzr = WordNetLemmatizer()
```

```python
df['text_string_lemm'] = df['text_string'].apply(lambda x:[lmtzr.lemmatize(word) for word in x])
df['text_string_lemm'] = df['text_string_lemm'].apply(lambda x : " ".join(x))
```

```python
df['text_string_lemm'] = df['text_string'].apply(lambda word: lmtzr.lemmatize(word, pos='v'))
```

```python
df.to_csv('clean_California_Senate_Sample_data.csv')
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

df['text_punct'] = df['text'].apply(lambda x: remove_punct(x))
df.head(10)
```
