```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.ipcc.ch/sr15/chapter/glossary/"
```

```python
response = requests.get(url)
```

```python
response = response.content
```

```python
response
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python
soup
```

```python
main = soup.find('main')
```

```python
articles = main.find_all('article', class_= 'chapter-section level-1')
```

```python
articles
```

```python
strong = articles.find_all('strong')
strong
```

```python
for tag in soup.find_all('strong'):
    print(tag.text)
```

```python
strong_all = soup.find_all('strong')
```

```python
keywords_list = []

# loop to extract just the company names from the strong tags, then using the a tags
    
for tag in strong_all:
    row = [tag.text]
    keywords_list.append(row)    
```

```python
keywords_list = pd.DataFrame(keywords_list, columns = ['keywords']).reset_index(drop = True)
keywords_list
```

```python
keywords_list.to_csv('keywords.csv')
```
